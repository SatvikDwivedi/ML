```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
df=pd.read_csv('/content/K-Means_assignment Data - Sheet1.csv')
```

```python
df.head(5)
```

```python
#euclidean method
dist_c1=[]
dist_c2=[]
dist_c3=[]
mins=[]
for i in df['Age']:
  dist_c1.append(math.sqrt(abs(6-i)))
  dist_c2.append(math.sqrt(abs(1-i)))
  dist_c3.append(math.sqrt(abs(10-i)))


df['cluster 1(6)']=dist_c1
df['Cluster 2 (1)']=dist_c2
df['cluster 3 (10)']=dist_c3

mean_c1=[]
mean_c2=[]
mean_c3=[]

for i in range(df['cluster 1(6)'].count()):
    if(dist_c1[i] > dist_c2[i]):
      if(dist_c3[i] > dist_c2[i]):
        mins.append(2)
        mean_c2.append(dist_c2[i])
      else:
        mins.append(3)
        mean_c3.append(dist_c3[i])
    else:
      mins.append(1)
      mean_c1.append(dist_c1[i])


df['Euclid_Clustering']=mins




print("The new cluster are C'1: ",sum(mean_c1)/len(mean_c1))
print("The new cluster are C'2: ",sum(mean_c2)/len(mean_c2))
print("The new cluster are C'3: 10")

df.head(10)



  
  
 


```

```python
#Manhattan method
distc1=[]
distc2=[]
distc3=[]
mins=[]
for i in df['Age']:
  distc1.append((abs(6-i)))
  distc2.append((abs(1-i)))
  distc3.append((abs(10-i)))


df['cluster 1(6)']=distc1
df['Cluster 2 (1)']=distc2
df['cluster 3 (10)']=distc3

meanc1=[]
meanc2=[]
meanc3=[]

for i in range(df['cluster 1(6)'].count()):
    if(distc1[i] > distc2[i]):
      if(distc3[i] > distc2[i]):
        mins.append(2)
        meanc2.append(distc2[i])
      else:
        mins.append(3)
        meanc3.append(distc3[i])
    else:
      mins.append(1)
      meanc1.append(distc1[i])



df['Manhattan_Clustering']=mins
df.head(10)
```
